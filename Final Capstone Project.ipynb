## 2. Performing Customers/sellers Segmentation

### a. Divide the customers into groups based on the revenue generated

```python
A = ORDERS
A = A.merge(CUSTOMERS, on = 'customer_id', how = 'left')
A = A.merge(ORDER_PAYMENTS, on = 'order_id', how = 'left')
B = A.loc[:,['customer_unique_id','payment_value']]
B['Deciles'] = pd.qcut( B.payment_value, 10, labels = False)
B.groupby('Deciles')[['customer_unique_id','payment_value']].agg( {'payment_value' : [np.min, np.max, sum]} ).reset_index().sort_values(by = 'Deciles', ascending = False)
```

### b. Divide the sellers into groups based on the revenue generated 

```python
A = ORDER_ITEMS
A = A.merge(SELLERS, on = 'seller_id', how = 'left')
B = A.loc[:,['seller_id','price']]
B['Deciles'] = pd.qcut( B.price, 10, labels = False)
B.groupby('Deciles')[['seller_id','price']].agg( {'price' : [np.min, np.max, sum]} ).reset_index().sort_values(by = 'Deciles', ascending = False)
```

## 3. Cross-Selling (Which products are selling together)
#### Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)

```python
from itertools import combinations
from collections import Counter
A = ORDER_ITEMS[ORDER_ITEMS['order_id'].duplicated(keep = False)]
A['Bundle'] = A.groupby('order_id')['product_id'].transform(lambda x: ','.join(x))
A = A[['product_id','Bundle']].drop_duplicates()
countx = Counter()
for row in A['Bundle']:
    row_list = row.split(',')
    countx.update(Counter(combinations(row_list,2)))
print(countx)
```

```python
# Printing top 10 among the cross selling products
countx.most_common(10)
```
