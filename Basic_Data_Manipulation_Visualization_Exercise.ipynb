### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
A = users.groupby(by=["occupation","gender"])[["gender"]].count().add_prefix("CO_").reset_index();
B = A.pivot(index="occupation",columns="gender",values="CO_gender");
B = B.fillna(0);
C = B.F + B.M;
D = B.M/C;
Result1 = D.sort_values(ascending=False);
print(Result1);
```

### 31. For each occupation, calculate the minimum and maximum ages

```python
agg_age = pd.concat([users.groupby('occupation')[['age']].max().add_prefix("Max_"), users.groupby('occupation')[['age']].min().add_prefix("Min_")], axis = 1)
agg_age
```

### 32. For each combination of occupation and gender, calculate the mean age

```python
avg_age = users.groupby(['occupation','gender'])[['age']].mean().add_prefix("Avg_")
avg_age
```

### 33.  For each occupation present the percentage of women and men

```python
A = users.groupby(by=["occupation","gender"])[["gender"]].count().add_prefix("CO_").reset_index();
B = A.pivot(index="occupation",columns="gender",values="CO_gender");
B = B.fillna(0);
C = B.F + B.M;
D = (B.M/C*100);
E = (B.F/C*100);
F = pd.concat([D,E], axis = 1);
F
```
